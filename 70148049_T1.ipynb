# Ejercicio 3

Escriba un programa que solicite 2 números y el símbolo de una operación aritmética e imprima un mensaje con el resultado de la operación aritmética entre los 2 números ingresados; en caso de que no exista el símbolo muestre un mensaje de error:

- Entrada: 1,2,+ -> Salida: La suma de 1 + 2 es 3
- Entrada: 2,1,- -> Salida: La resta de 2 - 1 es 1
- Entrada: 1,2,* -> Salida: La multiplicación de 1 * 2 es 2
- Entrada: 2,1,/ -> Salida: La división de 2 / 1 es 2

````python
try:
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    operacion = input("Ingrese el símbolo (+, -, *, /): ")
    if operacion == "+":
        print(f"La suma de {num1} + {num2} es {num1 + num2}")
    elif operacion == "-":
        print(f"La resta de {num1} - {num2} es {num1 - num2}")
    elif operacion == "*":
        print(f"La multiplicación de {num1} * {num2} es {num1 * num2}")
    elif operacion == "/":
        if num2 != 0:
            print(f"La división de {num1} / {num2} es {num1 / num2}")
        else:
            print("Error: No se puede dividir entre cero.")
    else:
        print("Error: El símbolo de operación no existe.")
except ValueError:
    print("Ingresar valores  válidos")
````

# Ejercicio 4

Escribir un programa que solicite las medidas de los lados de un triángulo (a, b, c) e imprima si es un triágulo escaleno, isósceles o equilátero.

````python
try:
    a = float(input("Ingrese el lado a: "))
    b = float(input("Ingrese el lado b: "))
    c = float(input("Ingrese el lado c: "))
    if a <= 0 or b <= 0 or c <= 0:
        print("Error: Los lados deben ser mayores a cero.")
    elif a + b <= c or a + c <= b or b + c <= a:
        print("Error: Los lados no forman un triángulo válido.")
    elif a == b == c:
        print("El triángulo es equilátero.")
    elif a == b or b == c or a == c:
        print("El triángulo es isósceles.")
    else:
        print("El triángulo es escaleno.")
except ValueError:
    print("Ingrese  valores numéricos")
````

# Ejercicio 5

Escriba un programa que solicite un $n$ y calcule la suma de la serie: $1^3 + 2^3 + 3^3 + ... + n^3$, donde $n$ es un entero positivo. Probar con $n=21$.

````python
try:
    n = int(input("Ingrese un número entero positivo n: "))
    if n <= 0:
        print("Error: n debe ser un entero positivo.")
    else:
        suma = 0
        for i in range(1, n + 1):
            suma += i ** 3
        print(f"La suma de 1³ + 2³ + ... + {n}³ es: {suma}")
except ValueError:
    print("Ingrese un número entero")
````

# Ejercicio 7

Escriba un programa que encuentre los números perfectos entre $1$ y $1000$, para hallar un número perfecto debe programar una función $perfecto(n)$ que retorne la suma de todos sus divisores enteros menores que $n$.

Ejemplo: $ 28 = 1 + 2 + 4 + 7 + 14 $


````python
def perfecto(n):
    suma = 0
    for i in range(1, n):
        if n % i == 0:
            suma += i
    return suma == n

for n in range(1, 1001):
    if perfecto(n):
        print(n, "es un número perfecto")
````

Escriba un programa con dos funciones, la primera obtendrá el módulo de un punto en el plano cartesiado, la segunda calculará la distancia entre dos puntos del plano cartesiano.

```python
# Ejemplo
a = (3, 4)
b = (7, 10)
print("Modulo de 'a':", modulo(a))
print("Distancia de 'a' hacia 'b':", distancia(a,b))
```

````python
def modulo(p):
    return (p[0]**2 + p[1]**2) ** 0.5

def distancia(p1, p2):
    return ((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2) ** 0.5

try:
    x1 = float(input("Ingrese x de 'a': "))
    y1 = float(input("Ingrese y de 'a': "))
    x2 = float(input("Ingrese x de 'b': "))
    y2 = float(input("Ingrese y de 'b': "))

    a = (x1, y1)
    b = (x2, y2)

    print("Modulo de 'a':", modulo(a))
    print("Distancia de 'a' hacia 'b':", distancia(a, b))

except ValueError:
    print("Ingrese solo valores numéricos")
````

# Ejercicio 9

Escriba un programa que lea un número entero positivo **n** y que a continuación imprima un cuadrado construido con asteriscos (*) de **n** asteriscos por lado.

Recuerde que la función **print** imprime por defecto un cambio de línea al final, si no quiere tener ese efecto deberá usar la función indicando el parámetro **end=""** por ejemplo al usar la función de este modo **print("*", end="")** se imprimirá el asterisco sin un cambio de línea al final. A continuación puede observar algunos resultados esperados de este ejercicio. Recuerde también que puede usar el carácter de " " espacio en blanco para generar el relleno del cuadrado que aparenta estar vacío.

| Entrada | Salida |
| :--- | :--- |
| 1 | `*` |
| 2 | `**`<br>`**` |
| 3 | `***`<br>`* *`<br>`***` |
| 4 | `****`<br>`*  *`<br>`*  *`<br>`****` |

````python
try:
    n = int(input("Ingrese n: "))
    if n <= 0:
        print("Error: n debe ser un entero positivo.")
    else:
        for i in range(n):
            for j in range(n):
                if i == 0 or i == n-1 or j == 0 or j == n-1:
                    print("*", end="")
                else:
                    print(" ", end="")
            print()
except ValueError:
    print("Ingrese solo un número entero")
````

# Ejercicio 10

Escriba un programa para calcular los promedios de una lista de estudiantes; el programa pide el nombre del estudiante seguido de su calificación; los nombres son ingresados en cualquier orden;
el ingresar un nombre vacío finaliza el ingreso de los datos;
una lista con todos los nombres y el promedio de cada estudiante debe ser mostrada al final.



```
# Ejemplo:
Ingresa el nombre del estudiante: Juan
Ingresa la calificación del estudiante (0-10): 7
Ingresa el nombre del estudiante: María
Ingresa la calificación del estudiante (0-10): 3
Ingresa el nombre del estudiante: Juan
Ingresa la calificación del estudiante (0-10): 2
Ingresa el nombre del estudiante: María
Ingresa la calificación del estudiante (0-10): 10
Ingresa el nombre del estudiante: María
Ingresa la calificación del estudiante (0-10): 3
Ingresa el nombre del estudiante: Juan
Ingresa la calificación del estudiante (0-10): 9
Ingresa el nombre del estudiante:
María : 5.333333333333333
Juan : 6.0
```

````python
estudiantes = {}

while True:
    nombre = input("Ingresa el nombre del estudiante: ")
    if nombre == "":
        break
    calificacion = float(input("Ingresa la calificación del estudiante (0-10): "))
    if nombre in estudiantes:
        estudiantes[nombre].append(calificacion)
    else:
        estudiantes[nombre] = [calificacion]

for nombre in estudiantes:
    promedio = sum(estudiantes[nombre]) / len(estudiantes[nombre])
    print(nombre, ":", promedio)
````
